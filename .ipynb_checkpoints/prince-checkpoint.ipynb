```python
def get_hreflists():
    url = 'https://gamewith.tw/pricone-re/article/show/87390'
    res = requests.get(url)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text,'html5lib')
    
    existchar_soup = soup.find_all('div',{'class':'puri_chara'})[1]
    notexistchar_soup = soup.find_all('div',{'class':'puri_chara'})[2]
    
    existchar_hreflist = [a['href'] for a in existchar_soup.find_all('a')]
    notexistchar_hreflist= [a['href'] for a in notexistchar_soup.find_all('a')]
    
    return [existchar_hreflist,notexistchar_hreflist]
```

```python
def getskillcycle(soup):
    if soup.find_all('div',{'class':'puri_pattern-table'})[0].find_all('tr')[0].get_text() != '循環':
        skill_start = soup.find_all('div',{'class':'puri_pattern-table'})[0].find_all('tr')[0].get_text()
        skill_re = soup.find_all('div',{'class':'puri_pattern-table'})[0].find_all('tr')[2].get_text()
        attack_start = soup.find_all('div',{'class':'puri_pattern-table'})[0].find_all('td')[0].find_all('img')
        attack_start = [str(img['src'])[-5].replace('r','◆') for img in attack_start]
        t = ''
        for i in attack_start:
            t+=' '+i
        attack_start = t[1:]

        attack_re = soup.find_all('div',{'class':'puri_pattern-table'})[0].find_all('td')[1].find_all('img')
        attack_re = [str(img['src'])[-5].replace('r','◆') for img in attack_re]
        t = ''
        for i in attack_re:
            t+=' '+i
        attack_re = t[1:]


        skill_cycle = pd.DataFrame([[attack_start,attack_re]],columns=[skill_start,skill_re])
        skill_cycle['name']=name
        skill_cycle.set_index('name',inplace=True)
    else:
        skill_re = soup.find_all('div',{'class':'puri_pattern-table'})[0].find_all('tr')[0].get_text()
        attack_re = soup.find_all('div',{'class':'puri_pattern-table'})[0].find_all('td')[0].find_all('img')
        attack_re = [str(img['src'])[-5].replace('r','◆') for img in attack_re]
        t = ''
        for i in attack_re:
            t+=' '+i
        attack_re = t[1:]
        
        skill_cycle = pd.DataFrame([['無',attack_re]],columns=['起手','起手後的循環'])
        skill_cycle['name']=name
        skill_cycle.set_index('name',inplace=True)
    return skill_cycle
```

```python
def get_table10(soup):
    hlist = soup.find_all(['h3','h4'])
    hdic={}
    history_h3 = ''
    history_h4 = ''
    for h in hlist:
        if str(h)[1:3] == 'h3' and history_h4 != '':
            hdic[history_h3] = history_h4
            history_h3 = ''
            history_h4 = ''

        if str(h)[1:3] == 'h3':
            h3 = h.get_text()
            hdic[h3]=''
            history_h3 = h3[-7:-5]
        if str(h)[1:3] == 'h4':
            h4 = h.get_text()
            history_h4 += h4

    values = [i[1] for i in hdic.items() if i[0] in ['優勢','弱點','用法']]
    col = [i[0] for i in hdic.items() if i[0] in ['優勢','弱點','用法']]
    table10 = pd.DataFrame([values],columns=col)
    table10['name']=name
    table10.set_index('name',inplace=True)
    return table10
```

```python
TorF=['T','F']
alldata = pd.DataFrame()
for publish in TorF:

    df = pd.DataFrame()
    col =[]
    if publish == 'T':
        hreflist = hreflists[0]
    elif publish == 'F':
        hreflist = hreflists[1]
#    else:
#        print("publish='T'or'F'")
    for href in hreflist:
        soup = get_soup(href)
        url = href

        nametext = soup.find_all('h1',{'class':'_title'})[0].get_text()
        name = re.findall(r'[\w\(\)]+',nametext)[0]
        table0 = pd.DataFrame(data=[name],columns=['name'])
        table0.set_index('name',inplace=True)


        table1 = pd.read_html(url,encoding='utf8',index_col=0)[0].T
        table1['name']=name
        table1.set_index('name',inplace=True)
        print((name,publish),end=',')


        table2 = pd.read_html(url,encoding='utf8',index_col=0)[1].T
        table2['name']=name
        table2.set_index('name',inplace=True)


        table4 = pd.read_html(url,encoding='utf8',index_col=0)[3].T
        try:
            table4.drop(index=2,inplace=True)
        except KeyError:
            table4.loc[1] = table4.loc[1].apply(lambda x:-1)
        table4.rename(columns={'物理防御力':'物理防禦力','魔法防御力':'魔法防禦力'},inplace=True)
        table4['name']=name
        table4.set_index('name',inplace=True)


        table5 = pd.read_html(url,encoding='utf8',index_col=0)[4].T
        table6 = pd.read_html(url,encoding='utf8',index_col=0)[5].T
        table7 = pd.read_html(url,encoding='utf8',index_col=0)[6].T
        table8 = pd.read_html(url,encoding='utf8',index_col=0)[7].T
        table9 = pd.read_html(url,encoding='utf8',index_col=0)[10].T


        skill = pd.concat([table5,table6,table7,table8]).T
        skill.columns=['必殺技','技能1','技能2','技能4']
        skill['name']=name
        skill.set_index('name',inplace=True)

        try:
            skill_cycle = getskillcycle(soup)
        except IndexError:
            print(name)

        table10 = get_table10(soup)


        dff = pd.concat([table0,table1,table2,table4,skill,skill_cycle,table10],ignore_index=False,axis=1,sort=False)
        if len(col)==0:
            col = dff.columns
            col = col.insert(0,'name')
        df = df.append(dff)
    #    print(dff)
    df.reset_index(inplace=True)
    df = df[col]
    data = df.copy()

    s = data['類型'].str.split(' / ',expand=True)
    s.columns=['位置','攻擊']
    df.drop(columns='類型',inplace=True)
    n=0
    for i in s.iteritems():
        df.insert(data.columns.get_loc('定位')+1+n,column=i[0],value=i[1])
        n+=1
    data = df.copy()

    s = data['技能4'].str.split('【EX技能\+】',expand=True)
    s.columns=['EX技能','EX+技能']
    df.drop(columns='技能4',inplace=True)
    n=0
    for i in s.iteritems():
        df.insert(data.columns.get_loc('技能4')+n,column=i[0],value=i[1])
        n+=1

    #data = data[['name', '評等分數', '定位', '位置','攻擊', '簡易評價', '初期★數', '種族', '公會', '声優', '記憶碎片',
    #       'HP', '物理攻撃力', '魔法攻撃力', '物理防禦力', '魔法防禦力', '必殺技', '技能1', '技能2', 'EX技能','EX+技能',
    #       '起手', '起手後的循環', '優勢', '弱點', '用法']]
    df.rename(columns={'name':'角色名稱'})
    df['評等分數'] = df['評等分數'].apply(lambda x:int(float(x.split(' /')[0])*10))
    df['必殺技'] = df['必殺技'].apply(lambda x:x.replace('【説明】',''))
    df['技能1'] = df['技能1'].apply(lambda x:x.replace('【説明】',''))
    df['技能2'] = df['技能2'].apply(lambda x:x.replace('【説明】',''))
    df['EX技能'] = df['EX技能'].apply(lambda x:x.replace('【説明】',''))
    df['實裝']=publish

    alldata = alldata.append(df)
    alldata.columns=df.columns
```
